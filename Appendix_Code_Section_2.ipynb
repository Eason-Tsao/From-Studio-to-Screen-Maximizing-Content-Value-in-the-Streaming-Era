# **Genre Trends over Years**

```python
# Convert 'YearMonth' column to datetime format
combined_catalogues['YearMonth'] = pd.to_datetime(combined_catalogues['YearMonth'])

# Group data by 'YearMonth' and 'Primary Genre' and count the occurrences
genre_trends_over_time = combined_catalogues.groupby(['YearMonth', 'Primary Genre']).size().reset_index(name='Frequency')

# Plotting the trends over time
plt.figure(figsize=(12, 8))
for genre in genre_trends_over_time['Primary Genre'].unique():
    genre_data = genre_trends_over_time[genre_trends_over_time['Primary Genre'] == genre]
    plt.plot(genre_data['YearMonth'], genre_data['Frequency'], label=genre)

plt.title('Genre Trends Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Frequency')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the frequency of each primary genre
genre_counts = combined_catalogues['Primary Genre'].value_counts()

# Get the top 5 primary genres
top_5_genres = genre_counts.head(5).index

# Filter the dataset to include only the top 5 primary genres
top_5_genre_data = combined_catalogues[combined_catalogues['Primary Genre'].isin(top_5_genres)]

# Group the filtered data by 'YearMonth' and 'Primary Genre' and count the occurrences
genre_trends_over_time = top_5_genre_data.groupby(['YearMonth', 'Primary Genre']).size().reset_index(name='Frequency')

# Plotting the trends over time for the top 5 primary genres
plt.figure(figsize=(12, 8))
for genre in top_5_genres:
    genre_data = genre_trends_over_time[genre_trends_over_time['Primary Genre'] == genre]
    plt.plot(genre_data['YearMonth'], genre_data['Frequency'], label=genre)

plt.title('Genre Trends Over Time (Top 5 Primary Genres)')
plt.xlabel('Year-Month')
plt.ylabel('Frequency')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np

# Define different shades of blue
blue_colors = ['#1f77b4', '#3182bd', '#6baed6', '#9ecae1', '#c6dbef']

# Group the data by 'Year' and 'Primary Genre' and sum up the frequency for each combination
genre_trends_over_time = combined_catalogues.groupby(['Year', 'Primary Genre']).size().reset_index(name='Frequency')

# Get unique years and top 5 genres
unique_years = genre_trends_over_time['Year'].unique()

# Sort the top 5 genres based on their frequency
top_5_genres_sorted = genre_trends_over_time.groupby('Primary Genre')['Frequency'].sum().nlargest(5).index.tolist()

# Plotting the trends over time for each genre
plt.figure(figsize=(12, 8))

# Width of each bar
bar_width = 0.15

# Initialize position for bars
bar_positions = np.arange(len(unique_years))

# Iterate over each genre
for i, genre in enumerate(top_5_genres_sorted):
    genre_data = genre_trends_over_time[genre_trends_over_time['Primary Genre'] == genre]
    frequencies = []

    for year in unique_years:
        if year in genre_data['Year'].values:
            frequencies.append(genre_data[genre_data['Year'] == year]['Frequency'].values[0])
        else:
            frequencies.append(0)

    # Shift position for bars to avoid overlap
    shifted_positions = bar_positions + (i - 2) * bar_width

    # Plot the bars for each genre with different shades of blue
    plt.bar(shifted_positions, frequencies, width=bar_width, label=genre, color=blue_colors[i])

    # Plot the trend line with the same color as the bars
    plt.plot(shifted_positions, frequencies, marker='o', linestyle='--', color=blue_colors[i])

plt.title('Genre Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(bar_positions, unique_years, rotation=15)
plt.legend(title='Primary Genre')
plt.tight_layout()
plt.grid(False)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'combined_catalogues' with columns 'Year' and 'Primary Genre'

# Step 1: Group the data by year and primary genre and count the occurrences
genre_counts_by_year = combined_catalogues.groupby(['Year', 'Primary Genre']).size().unstack()

# Step 2: Calculate the growth rate for each genre by comparing counts in consecutive years
growth_rates = genre_counts_by_year.pct_change() * 100

# Step 3: Plot the growth rates over years for the top genres
top_genres = genre_counts_by_year.sum().nlargest(5).index  # Assuming you want the top 5 genres

# Custom colors for the plot
custom_colors = ['#1f77b4', '#3182bd', '#6baed6', '#9ecae1', '#c6dbef']

# Plotting
growth_rates[top_genres].plot(figsize=(10, 6), marker='o', color=custom_colors)
plt.title('Growth Rate of Top Genres Over Years')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.legend(title='Primary Genre')
plt.grid(False)
plt.show()
```

**Genre Trends & Growth Rate:** The popularity of the Documentary genre has been consistently high since 2017, experiencing significant growth rates, particularly in 2017 and 2018.
Although the Romance genre has the lowest count among the top 5 genres, it was growing over years and exhibited higher growth rates compared to other genres in 2019 and 2020, indicating potential for future growth.

```python
import pandas as pd

# Convert 'Number of Episodes' to numeric
filtered_data['Number of Episodes'] = pd.to_numeric(filtered_data['Number of Episodes'], errors='coerce')

# Convert 'Duration' to numeric and coerce errors to NaN
filtered_data['Duration'] = pd.to_numeric(filtered_data['Duration'], errors='coerce')

# Initialize the 'duration_TVSeason' column with the 'Duration' column
filtered_data['duration_TVSeason'] = filtered_data['Duration']

# Perform the operation only for rows where content type is 'TVSeason' and both 'Duration' and 'Number of Episodes' are valid numeric values
filtered_data.loc[(filtered_data['Content Type'] == 'TVSeason') &
                  (~filtered_data['Duration'].isnull()) &
                  (~filtered_data['Number of Episodes'].isnull()),
                  'duration_TVSeason'] = filtered_data['Duration'] / filtered_data['Number of Episodes']


```

```python
# Define a function to calculate summary statistics for movies
def calculate_movie_stats(df):
    return df[df['Content Type'] == 'Movie']['Duration'].describe()

# Define a function to calculate summary statistics for TV seasons
def calculate_tvseason_stats(df):
    return df[df['Content Type'] == 'TVSeason']['duration_TVSeason'].describe()

# Group the data by 'Primary Genre' and apply the functions to calculate summary statistics
movie_stats = filtered_data.groupby('Primary Genre').apply(calculate_movie_stats)
tvseason_stats = filtered_data.groupby('Primary Genre').apply(calculate_tvseason_stats)

# Combine the results into a single DataFrame
summary_stats = pd.concat([movie_stats, tvseason_stats], axis=1, keys=['Movie', 'TVSeason'])

# Display the resulting summary statistics DataFrame
summary_stats
```

```python
median_stats = summary_stats.xs('50%', level=1, axis=1)
# Calculate the median duration for movies and TV seasons across the five different genres
median_avg_movie = median_stats['Movie'].mean()
median_avg_tvseason = median_stats['TVSeason'].mean()

# Plot the median durations for movies and TV seasons with median lines
plt.figure(figsize=(10, 6))
median_stats.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Median of Duration for Movies and TV Seasons by Genre')
plt.xlabel('Primary Genre')
plt.ylabel('Median of Duration (minutes)')
plt.xticks(rotation=15)
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0)

# Add median average lines for movies and TV seasons
plt.axhline(median_avg_movie, color='blue', linestyle='--', linewidth=1, label='Median Average (Movies)')
plt.axhline(median_avg_tvseason, color='green', linestyle='--', linewidth=1, label='Median Average (TV Seasons)')


plt.tight_layout()
plt.show()
```

**Content Duration Intrepretation**  
 Romance and Crime & Thriller genres have longer durations in both of the content types compared to other genres, while Children & Family are shorter, suggesting a preference for feature-length films in these categories. Studios can leverage these insights to diversify their content offerings, catering to audience preferences and maximizing viewer engagement across different genres and formats.
